```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from ydata_profiling import ProfileReport
```

```python
#get the data
data = pd.read_csv("NR_25001.csv")
                   
data.head(10)
```


2.- EDA

```python
# Using pandas.Series.dt.year()
#pd.to_datetime(data['FECHA'])
data['Year'] = pd.DatetimeIndex(data['FECHA']).year
data.head(10)
```

```python
data.head(10)  
```

```python
#Crear dataframe de huecos de datos por columnas
df_temp = data.copy()
fecha_col = 'FECHA'
columnas = 'PRECIP','EVAP','TMAX','TMIN'
df_temp = df_temp.sort_values(fecha_col)

resultados = []

for col in columnas:
    is_null = df_temp[col].isnull()
    es_inicio = is_null & (~is_null.shift(1, fill_value=False))
    group_id = es_inicio.cumsum()
    
    df_huecos = df_temp[is_null].copy()
    df_huecos['group_id'] = group_id[is_null]
    
    if df_huecos.empty:
        continue

    agg_results = df_huecos.groupby('group_id').agg(
        inicio=(fecha_col, 'min'),
        fin=(fecha_col, 'max')
    )
agg_results['duracion'] = (agg_results['fin'] - agg_results['inicio']).dt.days + 1    
    
    
    resultados.append(agg_results[['columna', 'inicio', 'fin', 'duracion']])
    
    if resultados:
        df_resultados = pd.concat(resultados, ignore_index=True)
    else:
        df_resultados = pd.DataFrame(columns=['columna', 'inicio', 'fin', 'duracion'])
        
total_duracion = df_resultados['duracion'].sum()   
print(f"Total duración de huecos {total_duracion} días")
```

```python
# Data preparation
dataset = data[['Year', 'PRECIP']].astype({'Year': 'string'}).groupby('Year').sum()
# Creating a line plot
dataset.set_index="Year"

dataset.plot(figsize = (15, 3), legend = True)
```

```python
# Data preparation
dataset_Evap = data[['Year', 'EVAP']].astype({'Year': 'string'}).groupby('Year').sum()
# Creating a line plot
dataset_Evap.set_index="Year"

dataset_Evap.plot(figsize = (15, 3), legend = True)
```

```python
# Data preparation
dataset_tmax = data[['Year', 'TMAX']].astype({'Year': 'string'}).groupby('Year').sum()
# Creating a line plot
dataset_tmax.set_index="Year"

dataset_tmax.plot(figsize = (15, 3), legend = True)
```
