## Proceamiento mensual total

```python
data = pd.read_csv('25033_mens.csv')                                          
data.head(10)
```

```python
#decomposition
decomposition = seasonal_decompose(x = data.EVAP,
                                   model = 'aditive',
                                   period = 12)
fig = decomposition.plot()
fig.set_size_inches((16, 9))
# Tight layout to realign things
fig.tight_layout()
plt.show()
plt.savefig("25033_evap_1961-2018_mensual.png", dpi=600)
```
