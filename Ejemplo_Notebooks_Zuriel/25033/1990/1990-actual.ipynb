```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Procesamiento de 1990 a la fecha diario

```python
#get the data
data = pd.read_csv('25033_90-2018.csv',
                   index_col='FECHA',
                   parse_dates=True)
data.tail(5)
```

```python
#decomposition
decomposition = seasonal_decompose(x = data.PRECIP,
                                   model = 'aditive',
                                   period = 365)
fig = decomposition.plot()
fig.set_size_inches((16, 9))
# Tight layout to realign things
fig.tight_layout()
plt.show()
plt.savefig("25033_precip_1990-2018.png", dpi=600)
```
