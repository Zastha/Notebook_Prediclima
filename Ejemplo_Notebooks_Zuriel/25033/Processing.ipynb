```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#get the data
data = pd.read_csv("25033.csv",
                   index_col='FECHA',
                   #parse_dates=True)
                   parse_dates=['FECHA_AUX'])
data.head(10)
```

```python
# Check the proportion of null values in the precip feature
precip_null_values = sum(data['PRECIP'].isnull() == True)
precip_notnull_values = sum(data['PRECIP'].isnull() == False)
print("PRECIP feature has {} null values and {} not null values !".format(precip_null_values, precip_notnull_values))
```

```python
#cleaning percip variable
import math
max_precip = max(data['PRECIP'])
min_precip = min(data['PRECIP'])
mean_precip = math.ceil(round(data["PRECIP"].mean(), 0))
median_precip = math.ceil(data['PRECIP'].median())
print("precip feature : max={}, min={}, mean={}, median={}".format(max_precip, min_precip, mean_precip, median_precip))
```

```python
#cleaning PRECIP variable
data['PRECIP'].fillna(value = mean_precip, inplace = True )
```

```python
#cleaning percip variable
import math
max_evap = max(data["EVAP"])
min_evap = min(data["EVAP"])
mean_evap = math.ceil(round(data["EVAP"].mean(), 0))
median_evap = math.ceil(data["EVAP"].median())
print("precip feature : max={}, min={}, mean={}, median={}".format(max_evap, min_evap, mean_evap, median_evap))
```

```python
data['EVAP'].fillna(value = mean_evap, inplace = True )
```

```python
# Check the proportion of null values in the EVAP feature
evap_null_values = sum(data['EVAP'].isnull() == True)
evap_notnull_values = sum(data['EVAP'].isnull() == False)
print("Evap feature has {} null values and {} not null values !".format(evap_null_values, evap_notnull_values))
```

```python
#cleaning TMAX variable
import math
max_tmax = max(data["TMAX"])
min_tmax = min(data["TMAX"])
mean_tmax = math.ceil(round(data["TMAX"].mean(), 0))
median_tmax = math.ceil(data["TMAX"].median())
print("Tmax feature : max={}, min={}, mean={}, median={}".format(max_tmax, min_tmax, mean_tmax, median_tmax))
```

```python
data['TMAX'].fillna(value = mean_tmax, inplace = True )
```

```python
# Check the proportion of null values in the EVAP feature
tmax_null_values = sum(data['TMAX'].isnull() == True)
tmax_notnull_values = sum(data['TMAX'].isnull() == False)
print("Tmax feature has {} null values and {} not null values !".format(tmax_null_values, tmax_notnull_values))
```

```python
#cleaning TMIN variable
import math
max_tmin = max(data["TMIN"])
min_tmin = min(data["TMIN"])
mean_tmin = math.ceil(round(data["TMIN"].mean(), 0))
median_tmin = math.ceil(data["TMIN"].median())
print("Tmin feature : max={}, min={}, mean={}, median={}".format(max_tmin, min_tmin, mean_tmin, median_tmin))
```

```python
data['TMIN'].fillna(value = mean_tmin, inplace = True )
```

```python
# Check the proportion of null values in the EVAP feature
tmin_null_values = sum(data['TMIN'].isnull() == True)
tmin_notnull_values = sum(data['TMIN'].isnull() == False)
print("Tmin feature has {} null values and {} not null values !".format(tmin_null_values, tmin_notnull_values))
```

```python
data['FECHA_AUX2']=pd.to_datetime(data['FECHA_AUX'], dayfirst=True, format='%d/%m/%Y')
data.head(5)
```

```python
data2=data.assign(Year=data["FECHA_AUX2"].dt.year,Month=data["FECHA_AUX2"].dt.month)
data2.tail(5)
```

```python
final.to_csv('25033_1990_2018_mens.csv', header=True, index=True) 
```
