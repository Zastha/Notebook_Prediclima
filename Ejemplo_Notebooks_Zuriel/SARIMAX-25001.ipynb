```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#get the data
data = pd.read_csv("25033_1990_2018_mens.csv",
                   index_col="FECHA",
                   parse_dates=True)
data.head(5)
```

```python
data['date'] = pd.to_datetime(data['FECHA2'],format='mixed')
data = data.set_index('date') 
data.PRECIP.resample('M').asfreq().ffill()
```

```python
#select variables

data.tail(10)  
```

```python
#viz
data['PRECIP'].plot(figsize = (10, 7), legend = True)
```

```python
#Stationarity
from pmdarima.arima  import ADFTest
stationarity = ADFTest(alpha=0.05)
stationarity.should_diff(data['PRECIP'])
```

#Preparation

```python
#Training and test set
test_ = 80
training_set = data.iloc[:-test_, :]
test_set = data.iloc[-test_:, :]
training_set.tail()
```

#SARIMAX

```python
#Exogenous VAriables
train_exog = training_set.iloc[:, 0:]
test_exog = test_set.iloc[:, 1:]
train_exog.tail(10)
```

```python
training_set['PRECIP']
```

```python
%pip install numpy==1.26.3
```

```python
#SARIMAX
from pmdarima import auto_arima
model = auto_arima(y = training_set["PRECIP"],
                   start_p=0, d=1,max_d=5,start_P=0,D=1,start_Q=0,max_P=5,max_D=5,max_Q=5,seasonal=True,
                   m = 12,error_action='warn',trace=True,suppress_warnings=True,random_state=20,n_fits=50,
                   stepwise = True)
model.summary()
```

```python
#Predictions
predictions = pd.Series(model.predict(n_periods= test_,
                        X = test_exog)).rename("SARIMAX")
predictions.index = test_set.index
predictions.head(10)
```

```python
#Plotting
training_set['PRECIP'].plot(legend = True,   figsize = (9,6))
test_set['PRECIP'].plot(legend = True)
predictions.plot(legend = True)
```

```python
#MAPE
def MAPE(y_true, y_pred):
  y_true, y_pred = np.array(y_true), np.array(y_pred)
  return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
MAPE(test_set.Churrasco, predictions)
```
